# STRUCTURE VALIDATION EXTRACTION

This notebook contians the necessary code to get the validation reports from the 3D structures in our dataset.

```python
import pandas as pd
import numpy as np
import os
import structure_validation
import importlib
```

```python
df = pd.read_pickle('/cluster/gjb_lab/2394007/varalign_runs/varalign_bio_units4/.varalign/prointvar_analysis_data/anks_final_aln.sto_prointvar_structure_table.p.gz') # STRUCTURAL DATA TABLE
```

```python
print(len(df))
```

```python
validation_dir = "/cluster/gjb_lab/2394007/structure_validation"
pdb_dir = "/cluster/gjb_lab/2394007/varalign_runs/varalign_bio_units3/.prointvar/pdbx"
```

```python
structure_validation.download_validation(df, validation_dir) # DOWNLOADS VALIDATION DATA USING PROTEOFAV
rs_dict = structure_validation.get_rsrz_dict(df, validation_dir) # GETS DICT INDICATING WHICH PDBs HAVE RSRZ ON THEIR VALIDATION
struc_df = structure_validation.get_res_exp_from_df(df, pdb_dir) # GETS RESOLUTION AND EXPERIMENTAL METHOD USED FOR SOLVING EACH PDB
val_df = structure_validation.get_validation_df(df, validation_dir) # CONCATENATES ALL VALIDATION TABLES IN A BIG DF
validation_def = pd.merge(val_df, struc_df, on = "pdb_id") # MERGES VALIDATION DF WITH RESOLUTION AND EXP. DATA DF
merged_df = structure_validation.merge_with_validation(df, validation_def) # MERGES PREVIOUS DF WITH STRUCTURAL TABLE DATA
```
